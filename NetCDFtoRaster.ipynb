```python
import arcpy  
from arcpy import env  
from arcpy.sa import *
import glob
```

```python
arcpy.env.workspace = "F:/MonthlyComp/2019" 
arcpy.env.overwriteOutput = True
```

```python
OutputFolder = "F:/MonthlyComp/2019_R"
```

```python
NCfiles = glob.glob('F:/MonthlyComp/2019/*.nc')
```

```python
print(NCfiles)
```

```python
for filename in NCfiles:
    if "baa" in filename:
        print("Processing: " + filename)  
    #inNCfiles = arcpy.env.workspace + "/" + filename
        fileroot = filename.split('/')[2]
        fileroot1 = fileroot.split('\\')[1]
        fileroot2 = fileroot1.split('_')[1] + fileroot1.split('_')[3].split('.nc')[0]
        print (fileroot2)
        TempLayerFile = "bleaching_alert_area"  
        outRaster = OutputFolder + "/" + str(fileroot2)
        arcpy.MakeNetCDFRasterLayer_md(filename, "bleaching_alert_area", "lon", "lat", TempLayerFile, "", "", "BY_VALUE")  
        arcpy.CopyRaster_management(TempLayerFile, outRaster + ".tif", "", "", "", "NONE", "NONE", "")
    elif "dhw" in filename:
        print("Processing: " + filename)  
    #inNCfiles = arcpy.env.workspace + "/" + filename
        fileroot = filename.split('/')[2]
        fileroot1 = fileroot.split('\\')[1]
        fileroot2 = fileroot1.split('_')[1] + fileroot1.split('_')[3].split('.nc')[0]
        print (fileroot2)
        TempLayerFile = "degree_heating_week"  
        outRaster = OutputFolder + "/" + str(fileroot2)
        arcpy.MakeNetCDFRasterLayer_md(filename, "degree_heating_week", "lon", "lat", TempLayerFile, "", "", "BY_VALUE")  
        arcpy.CopyRaster_management(TempLayerFile, outRaster + ".tif", "", "", "", "NONE", "NONE", "")
    elif "hs" in filename:
        print("Processing: " + filename)  
    #inNCfiles = arcpy.env.workspace + "/" + filename
        fileroot = filename.split('/')[2]
        fileroot1 = fileroot.split('\\')[1]
        fileroot2 = fileroot1.split('_')[1] + fileroot1.split('_')[3].split('.nc')[0]
        print (fileroot2)
        TempLayerFile = "hotspot"  
        outRaster = OutputFolder + "/" + str(fileroot2)
        arcpy.MakeNetCDFRasterLayer_md(filename, "hotspot", "lon", "lat", TempLayerFile, "", "", "BY_VALUE")  
        arcpy.CopyRaster_management(TempLayerFile, outRaster + ".tif", "", "", "", "NONE", "NONE", "")
    elif "ssta" in filename:
        print("Processing: " + filename)  
    #inNCfiles = arcpy.env.workspace + "/" + filename
        fileroot = filename.split('/')[2]
        fileroot1 = fileroot.split('\\')[1]
        fileroot2 = fileroot1.split('_')[1] + fileroot1.split('_')[3].split('.nc')[0]
        print (fileroot2)
        TempLayerFile = "sea_surface_temperature_anomaly"  
        outRaster = OutputFolder + "/" + str(fileroot2)
        arcpy.MakeNetCDFRasterLayer_md(filename, "sea_surface_temperature_anomaly", "lon", "lat", TempLayerFile, "", "", "BY_VALUE")  
        arcpy.CopyRaster_management(TempLayerFile, outRaster + ".tif", "", "", "", "NONE", "NONE", "")
    elif "sst" in filename:
        print("Processing: " + filename)  
    #inNCfiles = arcpy.env.workspace + "/" + filename
        fileroot = filename.split('/')[2]
        fileroot1 = fileroot.split('\\')[1]
        fileroot2 = fileroot1.split('_')[1] + fileroot1.split('_')[3].split('.nc')[0]
        print (fileroot2)
        TempLayerFile = "sea_surface_temperature"  
        outRaster = OutputFolder + "/" + str(fileroot2)
        arcpy.MakeNetCDFRasterLayer_md(filename, "sea_surface_temperature", "lon", "lat", TempLayerFile, "", "", "BY_VALUE")  
        arcpy.CopyRaster_management(TempLayerFile, outRaster + ".tif", "", "", "", "NONE", "NONE", "")
    else:
        print('not processed: ', filename)
print ("Done")        
```
